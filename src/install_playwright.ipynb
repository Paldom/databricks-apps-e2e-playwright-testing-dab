```python
import shutil
import subprocess
import sys

def run(cmd: list[str]) -> None:
    print(f"+ {' '.join(cmd)}")
    subprocess.check_call(cmd)

playwright_bin = shutil.which("playwright")
playwright_cmd = [playwright_bin] if playwright_bin else [sys.executable, "-m", "playwright"]

# Install Chromium and required Linux shared libs (libatk, libgtk-3, etc.).
run([*playwright_cmd, "install", "--with-deps", "chromium"])

print("Playwright + Chromium + system dependencies installed successfully.")
```
